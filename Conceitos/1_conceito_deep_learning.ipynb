# Deep Learning – Conceitos Fundamentais

## 1. O que é uma Rede Neural Artificial (ANN)
"""
Uma rede neural artificial (ANN) é um modelo computacional inspirado no funcionamento do cérebro humano.
É composta por "neurônios artificiais" organizados em camadas: entrada, ocultas e saída.

Cada neurônio realiza uma combinação linear das entradas, aplica uma função de ativação e propaga o resultado.

Fórmula de um neurônio:
    y = f(Wx + b)
onde:
    - x: vetor de entrada
    - W: pesos
    - b: bias (viés)
    - f: função de ativação
"""

## 2. Perceptron Simples e Multicamadas (MLP)
"""
O perceptron simples é a unidade básica de uma rede neural, capaz apenas de aprender funções linearmente separáveis.

MLP (Multi-Layer Perceptron) adiciona camadas ocultas e funções de ativação não lineares, permitindo aprendizado de padrões complexos.

Estrutura típica:
Entrada → [Camada Oculta 1] → [Camada Oculta 2] → Saída

Utiliza o algoritmo de backpropagation para ajustar os pesos.
"""

## 3. Funções de Ativação

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-10, 10, 100)

sigmoid = 1 / (1 + np.exp(-x))
tanh = np.tanh(x)
relu = np.maximum(0, x)
softmax = np.exp(x) / np.sum(np.exp(x))  # Vetorialmente, apenas ilustrativo

plt.figure(figsize=(12, 8))
plt.subplot(2, 2, 1)
plt.plot(x, sigmoid)
plt.title('Sigmoid')

plt.subplot(2, 2, 2)
plt.plot(x, tanh)
plt.title('tanh')

plt.subplot(2, 2, 3)
plt.plot(x, relu)
plt.title('ReLU')

plt.subplot(2, 2, 4)
plt.plot(x, softmax)
plt.title('Softmax (ilustrativo)')

plt.tight_layout()
plt.show()
```

## 4. Forward Propagation e Backpropagation
"""
- Forward propagation: os dados percorrem a rede camada por camada até a saída.
- A saída é comparada com o valor real usando uma função de perda (loss).

- Backpropagation: calcula o gradiente do erro em relação aos pesos e os ajusta usando descida do gradiente.

Fluxo:
Entrada → Forward → Cálculo da perda → Backward → Ajuste de pesos

Principais funções de perda:
- MSE: regressão
- Categorical Crossentropy: classificação multiclasse
"""

## 5. Overfitting e Regularização
"""
Overfitting ocorre quando a rede aprende tão bem os dados de treino que perde a capacidade de generalizar.

Técnicas para reduzir overfitting:
- Dropout: desativa aleatoriamente neurônios durante o treino
- EarlyStopping: para o treino quando a performance no conjunto de validação para de melhorar
- Regularização L1/L2: penaliza pesos muito grandes

Exemplo em Keras:
model.add(Dense(64, activation='relu'))
model.add(Dropout(0.5))
"""
